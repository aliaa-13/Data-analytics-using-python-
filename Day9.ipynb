```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv('Sample - Superstore.csv', encoding='ISO-8859-1')

df.head()
df.info()
df.describe()

region_data = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
print(region_data)

cat_data = df.groupby('Category')['Sales'].sum()
print(cat_data)

subcat_data = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
print(subcat_data)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title('Discount vs Profit')
plt.grid(True)
plt.show()

loss_items = df[df['Profit'] < 0].groupby('Sub-Category')['Profit'].sum().sort_values().head(5)
print(loss_items)
```
